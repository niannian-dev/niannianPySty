# 学习 ndarray
## 1.ndarray特性
### 多维性

```python
import numpy as np
```

```python
arr = np.array(5)   #创建一个0维数组
print(arr)
print("数组的维度为:", arr.ndim)     #查看数组的维度
```

```python
arr = np.array([1,2,3])   #创建一个1维数组
print(arr)
print("数组的维度为:", arr.ndim)     #查看数组的维度
```

```python
arr = np.array([[1,2,3],[4,5,6]])   #创建一个2维数组
print(arr)
print("数组的维度为:", arr.ndim)     #查看数组的维度
```

### 同质性

```python
arr = np.array([1,'a'])     #将1转换成字符串，强制类型转换
print(arr)
```

```python
arr = np.array([1,1.5])     #将1转换成浮点数，强制类型转换
print(arr)
```
