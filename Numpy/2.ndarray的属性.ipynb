```python
import numpy as np
```

```python
arr0 = np.array(1)
print('数组的形状为:',arr0.shape)
print('数组的维度为:',arr0.ndim)
print('数组的元素个数为:',arr0.size)
print('数组的元素类型为:',arr0.dtype)
print('数组的转置为:',arr0.T)
print('数组的每个元素的字节数为:',arr0.itemsize)
print('数组的字节数为:',arr0.nbytes)
print('数组的标志位为:',arr0.flags)
```
