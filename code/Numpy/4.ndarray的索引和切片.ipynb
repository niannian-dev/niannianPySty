# 索引和切片

```python
import numpy as np
```

```python
arr = np.random.randint(1,1000,20)
print(arr)


print(arr[10])
print(arr[:])   # 打印数组的所有元素
print(arr[10:20])   # 打印数组的第11个元素到第20个元素 左闭右开
print(arr[::2])   # 打印数组的所有偶数索引元素
print(arr[slice(10,20,2)])   # 打印数组的第11个元素到第20个元素 左闭右开，步长为2
print(arr[(arr > 200) & (arr < 500)])   # 布尔索引
```

```python
# 二位数组的索引和切片
arr = np.random.randint(500,1000,(4,5))

print(arr)

print(arr[2,3]) # 打印数组的第3行第4列的元素
print(arr[1, :])    # 打印数组的第2行的所有元素
print(arr[:, 2])    # 打印数组的第3列的所有元素
print(arr[2][ arr[2] > 700])     # 打印数组的第3行大于700的元素
print(arr[:,2][ arr[:,2] > 700])     # 打印数组的第3列大于700的元素 
```
