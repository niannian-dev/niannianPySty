# ndarray的运算

```python
import numpy as np
```

```python
# 数组的加法
a1 = np.array([1,2,3])
a2 = np.array([4,5,6])
print(a1 + a2)
print(a1 - a2)
print(a1 * a2)
print(a1 / a2)
```

```python
# python原生的加法
a1 = [1,2,3]
a2 = [4,5,6]
a3 = []
for i in range(len(a1)):
    a3.append(a1[i] + a2[i])
print(a3)

# 原生的加法是拼接操作，没有减法、乘法、除法
print(a1 + a2)
```

```python
# 二维数组的运算    ，注意这不是矩阵的乘除法
a1 = np.array([[1,2,3],[4,5,6]])
a2 = np.array([[7,8,9],[10,11,12]])
print(a1 + a2)
print(a1 - a2)
print(a1 * a2)
print(a1 / a2)
```

```python
# 矩阵和标量的运算
a1 = np.array([[1,2,3],[4,5,6]])
print(a1 + 10)
print(a1 - 10)
print(a1 * 10)
print(a1 / 10)
```

```python
# 广播机制
a = np.array([[1],[2],[3]])
b = np.array([4,5,6])

'''
    1 2 3

    4
    5
    6
 
    1 2 3
    1 2 3
    1 2 3   

    4 4 4
    5 5 5
    6 6 6

'''

print(a + b)
```

```python
# 矩阵的乘法运算
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[4,5,6],[3,2,1]])
print(a @ b.T)
```
