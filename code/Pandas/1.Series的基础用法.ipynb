# series的创建

```python
import pandas as pd
```

```python
s1 = pd.Series([1,2,3,4,5])

print(s1)

s2 = pd.Series([1,2,3,4,5] , index = ['a','b','c','d','e'] )

print(s2)

s3 = pd.Series({'a':1,'b':2,'c':3,'d':4,'e':5})

print(s3)    

s4 = pd.Series(s3 , index = ['a','e','c'] )

print(s4)
```

```python
# Series的属性

s = pd.Series({'a':1,'b':2,'c':3,'d':4,'e':5})

# index
print(s.index)

# values
print(s.values)

# shape
print(s.shape)

# ndim
print(s.ndim)

# size
print(s.size)

# dtype
print(s.dtype)

# name
print(s.name)
s.name = 'test'
print(s.name)

# loc   显示索引
print(s.loc['a'])
# 支持切片
print(s.loc['a':'c'])

# iloc   隐式索引
print(s.iloc[3])
print(s.iloc[0:3])

# at 获取指定索引的值   不支持切片
print(s.at['a'])

# iloc 获取指定索引的值
print(s.iloc[3])
```

```python
s = pd.Series({'a':1,'b':2,'c':3,'d':4,'e':5,'f':6,'g':7,'h':8})
print(s)

# 访问索引
print(s['a'])   # 根据索引(显示索引)访问数据
print(s.iloc[0])   # 根据索引(隐式索引)访问数据
print(s[s<3])   # 布尔索引访问数据
```
