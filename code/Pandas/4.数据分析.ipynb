```python
# 缺失值的处理
import pandas as pd
import numpy as np

s = pd.Series([1,2,np.nan,4,5,None,pd.NA])
print(s)

h = pd.DataFrame([[None,2,3],[4,np.nan,6],[7,8,pd.NA]])
print(h)

print('1.判断缺失值')
print(s.isnull())
print('2.判断缺失值')
print(s.isna())
print('3.判断缺失值')
print(h.isnull())
print('4.判断缺失值')
print(h.isna())
```

```python
# 缺失值数量
print(h.isnull().sum())
print(s.isna().sum())
```

```python
# 剔除缺失值
print(s.dropna())
print(s)

# 只要有一个缺失值，就剔除整行
print(h.dropna())
print(h)

print(h.dropna(how='all'))  # 只有当整行都是缺失值时，才会剔除整行
print(h)
```

```python
# 根据条件剔除缺失值
h = pd.DataFrame([[None,pd.NA,3],[4,np.nan,6],[7,8,pd.NA]])

print(h.dropna(thresh=2))  # 只有当整行非缺失值的数量大于等于2时，才会剔除整行
print(h)
```

```python
# 删除一整列    
# axis  表示列方向上
print(h.dropna(axis=1,thresh=2))
print(h)
```

```python
# 根据列名剔除缺失值
h = pd.DataFrame([[None,pd.NA,3],[4,np.nan,6],[7,8,pd.NA]],columns=['a','b','c'])
print(h.dropna(subset=['a']))  # 只有当整行a列非缺失值的数量大于等于1时，才会剔除整行
print(h)
```

```python
# 填充缺失值
df = pd.read_csv('data/weather_withna.csv')
df.isna().sum()

# 使用字典填充缺失值，key为列名，value为填充值
print(df.fillna({'temp_max':20,'wind':2.5}).tail())

# 使用统计值填充缺失值
print(df.fillna(df[['wind']].mean()).tail())

# 根据缺失值周边数据填充缺失值
print(df.ffill().tail())    # 前向填充
print(df.bfill().tail())    # 后向填充
```

```python
# 数据变形
import pandas as pd
data = {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['张三', '李四', '王五', '赵六', '钱七'],
    'Math': [80, 90, 75, 95, 88],
    'English': [95, 85, 90, 88, 92],
    'Physics': [88, 92, 85, 90, 95],
   }


df = pd.DataFrame(data)
df
df.T    # 转置
```

```python
# 宽表转换成长表，将每个科目转换为一行，包含ID、姓名、科目和成绩四列
df2 = pd.melt(df, id_vars=['ID', 'Name'], var_name='Subject', value_name='Score')
df2.sort_values('Name') 
```

```python
# 长表转宽表
pd.pivot(df2,index=['ID','Name'],columns='Subject',values='Score')
```

```python
# 数据分箱  pd.cut(x,bins,labels)
# 将数据分割成多个区间
df = pd.read_csv('data/employees.csv')
df
```

```python
df1 = df.head(10)[['employee_id','salary']]

df1
```

```python
# 数据分箱
pd.cut(df1['salary'],bins=3)
pd.cut(df1['salary'],bins=3).value_counts()
```

```python
# 设定分箱区间
pd.cut(df1['salary'],bins=[0,10000,20000,30000])

df1['收入范围'] =pd.cut(df1['salary'],bins=[0,10000,20000,30000],labels=['低','中','高'])
df1
```

```python
# 等分
pd.qcut(df1['salary'],q=3).value_counts()
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/sleep.csv')
df1 = df.head()
df1 = df[['person_id','sleep_quality','gender']]
df1
```

```python
df1['睡眠质量'] = pd.cut(df1['sleep_quality'], bins=3,labels=['差','中','好'])
df1
```

```python
# 数据分箱只能用于数值型数据，如果要分类字符串类型等，可以设置字符串为category类型
print(df1['睡眠质量'].value_counts())

print(df1['gender'].astype('category'))
```
