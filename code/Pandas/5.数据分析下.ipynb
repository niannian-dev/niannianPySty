```python
import pandas as pd
```

```python
# 时间数据的处理
d = pd.Timestamp('2026-04-21')
print(d)
print(type(d))
print('年份:',d.year)
print('月份:',d.month)
print('日期:',d.day)

print('时间:',d.minute,d.second)
print('季度:',d.quarter)
print('是否是月底:',d.is_month_end)
print('星期几:',d.day_name())

print('转换为天数',d.to_period('D'))
print('转换为季度',d.to_period('Q'))
print('转换为年份',d.to_period('Y'))
print('转换为月份',d.to_period('M'))
print('转换为周',d.to_period('W'))
```

```python
# 字符串转换为日期
# d = pd.to_datetime('2023-01-01')
d = pd.to_datetime('2026-04-22T23:05:14.826+08:00')
print(d)
print(type(d))
print('年份:',d.year)
print('月份:',d.month)
print('日期:',d.day)
```

```python
import pandas as pd

df = pd.read_csv('data/employees.csv')

df
# 分组聚合
df.dropna(subset=['department_id'], inplace=True)

df['department_id'] = df['department_id'].astype('int64')

# 计算不同部门的平均工资

# 查看分组
df.groupby('department_id').groups

df.groupby('department_id').get_group(20)
df2 = df.groupby('department_id')[['salary']].mean()
df2['salary'] = df2['salary'].round(2)

print(df2)

df2 = df2.reset_index()

print(df2)

df2.sort_values('salary', ascending=False)

print(df2)
```

```python
import pandas as pd

df = pd.read_csv('data/employees.csv')

df
# 分组聚合
df.dropna(subset=['department_id'], inplace=True)

df['department_id'] = df['department_id'].astype('int64')

df2 = df.groupby(['department_id','job_id'])[['salary']].mean()
df2 = df2.reset_index()

df2['salary'] = df2['salary'].round(2)
df2 = df2.sort_values('salary', ascending=False)
print(df2)
```
