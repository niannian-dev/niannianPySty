```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.family'] = ['SimHei']
```

```python
# 折线图
# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 绘图数据
month = ['1月','2月','3月','4月','5月','6月','7月','8月','9月','10月','11月','12月']
sales = [1000, 1500, 1200, 1800, 1600, 1400, 1300, 1500, 1700, 1900, 2000, 2100]

# 绘制折线图
plt.plot(month, sales, label='产品A',marker='o',markersize=3)

# 添加标题
plt.title('月销售数据', fontsize=20,color='red')

# 添加坐标轴标签
plt.xlabel('月份')
plt.ylabel('销售量(万元)')

# 添加图例
# plt.legend(['销售量'])
plt.legend(loc='upper left')

# 添加网格线
plt.grid(True,alpha=0.3,linestyle='--')

# 设置刻度字体大小
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12,rotation=0.0)

# 设置y轴范围
plt.ylim(0, 2500)

# 在每个数据点上显示数值
for x,y in zip(month,sales):
    print(x,y)
    plt.text(x,y+20,str(y),ha='center',va='bottom')

# 显示图表
plt.show()
```

```python
# 条形图
# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 绘图数据
subjects = ['语文','数学','英语','物理','化学','生物','政治']
scores = [80,90,75,95,85,90,88]

# 绘制条形图
plt.bar(subjects, scores,label='张三',color='green',width=0.5)

# 添加标题
plt.title('成绩分布图', fontsize=20,color='blue')

# 添加坐标轴标签
plt.xlabel('科目')
plt.ylabel('成绩')

# 添加图例
# plt.legend(['销售量'])
plt.legend(loc='upper left')

# 添加网格线
plt.grid(axis='y',alpha=0.3,linestyle='--')

#设置刻度字体大小
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12,rotation=0.0)

# 设置y轴范围
plt.ylim(40, 100)

# 在每个数据点上显示数值
for x,y in zip(subjects,scores):
    print(x,y)
    plt.text(x,y+1,str(y),ha='center',va='bottom')

# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 柱状图
# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 绘图数据
countries = ['中国','美国','英国','法国','德国','意大利','西班牙']
gdp = [2433,2137,2830,2730,2630,2530,2430]

# 绘制条形图
plt.barh(countries, gdp)

# 添加标题
plt.title('国家GDP对比图', fontsize=20,color='blue')

# 添加坐标轴标签
plt.xlabel('国家')
plt.ylabel('GDP')

# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 饼图
# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 绘图数据
things = ['学习','工作','休息','运动']
times = [3,4,5,6]
colors = ['red','green','blue','yellow']

# 绘制饼图
plt.pie(times, labels=things,autopct='%1.1f%%',startangle=90,colors=colors)

# 添加标题
plt.title('时间分配图', fontsize=20,color='blue')

# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 饼图
# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 绘图数据
things = ['学习','工作','休息','运动']
times = [3,4,5,6]
colors = ['red','green','blue','yellow']

# 绘制饼图
plt.pie(times, labels=things,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        wedgeprops=dict(width=0.5),
        pctdistance=0.7
        )
# 添加标题
plt.title('时间分配图', fontsize=20,color='blue')
plt.text(-0.2, 0, '时间分配图',color='blue')

# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 爆炸饼图
# 创建图表，设置大小
plt.figure(figsize=(10, 6))

# 绘图数据
things = ['学习','工作','休息','运动']
times = [3,4,5,6]
colors = ['red','green','blue','yellow']
explode = (0.1, 0, 0, 0)

# 绘制饼图
plt.pie(times, labels=things,autopct='%1.1f%%',
        startangle=0,colors=colors,explode=explode)

# 添加标题
plt.title('时间分配图', fontsize=20,color='blue')

# 自动优化排版
plt.tight_layout()

# 显示图表
plt.show()
```

```python
# 散点图
# 创建图表，设置大小
plt.figure(figsize=(10, 8))
import random

# 绘图数据
x = []
y = []
for i in range(1,100):
    tmp = random.uniform(0,10)
    x.append(tmp)
    y.append(tmp*2 + random.gauss(0,2))

# # 绘制折线图
plt.scatter(x, y,
            color='red',
            alpha=0.5,
            s=10,
            )

# # 添加标题
plt.title('x-y散点图', fontsize=20,color='red')

# # 添加坐标轴标签
plt.xlabel('x')
plt.ylabel('y')

# # 添加图例
plt.legend(['x-y散点图'])
# plt.legend(loc='upper left')

# # 添加网格线
plt.grid(True,alpha=0.3,linestyle='--')

# # 设置刻度字体大小
plt.xticks(fontsize=12,rotation=45)
plt.yticks(fontsize=12,rotation=0.0)

# # 设置y轴范围
plt.ylim(0, 30)
plt.plot([0,10],[0,20],color='orange')

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 设置随机种子，保证结果可复现
np.random.seed(42)

# 生成 20 名学生的三科成绩（分数范围 0~100）
chinese = np.random.normal(75, 12, 20)      # 语文：均值75，标准差12
math = np.random.normal(68, 15, 20)         # 数学：均值68，标准差15
english = np.random.normal(80, 10, 20)      # 英语：均值80，标准差10

# 人为添加异常值，增强箱线图特征
chinese = np.append(chinese, [25, 98])      # 语文添加低分和高分
math = np.append(math, [100, 42])           # 数学添加高分和低分
english = np.append(english, [55, 96])      # 英语添加低分和高分

# 将成绩限制在 [0,100] 范围内（避免随机生成超出边界的值）
chinese = np.clip(chinese, 0, 100)
math = np.clip(math, 0, 100)
english = np.clip(english, 0, 100)

# 数据准备：三个列表，每个对应一科成绩
data = [chinese, math, english]
labels = ['语文', '数学', '英语']

# 绘制箱线图
plt.figure(figsize=(8, 6))
box = plt.boxplot(data, labels=labels, patch_artist=True,
                  showmeans=True, meanline=True,  # 显示均值线
                  medianprops={'color': 'red', 'linewidth': 1.5},
                  meanprops={'color': 'blue', 'linestyle': '--', 'linewidth': 1.2},
                  whiskerprops={'linestyle': '-', 'color': 'gray'},
                  capprops={'color': 'black'},
                  flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 6, 'alpha': 0.6})

# 填充箱体颜色
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# 添加标题和轴标签
plt.title('20名学生语数英三科成绩箱线图', fontsize=14, pad=15)
plt.ylabel('分数', fontsize=12)
plt.xlabel('科目', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 显示图形
plt.tight_layout()
plt.show()
```
