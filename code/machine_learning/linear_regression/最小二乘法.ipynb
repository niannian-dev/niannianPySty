```python
import pandas as pd


data = pd.read_csv('data.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
data.plot.scatter(x='study_hours', y='score')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
features = data['study_hours'].values.reshape(-1, 1)  # 特征,x
target = data['score']  # 目标,y

regression = LinearRegression()
model = regression.fit(features, target)
```

```python
# 最小二乘法，截距，b
model.intercept_
```

```python
# 最小二乘法，斜率，k
model.coef_
```

```python
# 根据特征预测目标
model.predict([[6]])
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.family'] = ['SimHei']

from sklearn.linear_model import LinearRegression
features = data['study_hours'].values.reshape(-1, 1)  # 特征,x
target = data['score']  # 目标,y

# 创建一个线性回归模型对象
regression = LinearRegression()
model = regression.fit(features, target)

ax = data.plot.scatter(x='study_hours', y='score',label="实际数据")
ax.plot(features, model.predict(features), color='red',label=f"拟合线:  y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}")

plt.legend()

plt.show()
```
